```python
import matplotlib.pyplot as plt
import numpy as np

# data
x = np.array([0, 100])
y = np.array([0, 10])

# slope/intercept for the line through the two points
m = (y[1] - y[0]) / (x[1] - x[0])
b = y[0] - m * x[0]

# dense x values for plotting the regression line
x_line = np.linspace(0, 100, 500)
y_line = m * x_line + b

fig, ax = plt.subplots()
ax.set_xlim(0, 100)
ax.set_ylim(0, 10)

x_positions = [0, 10, 80, 100]
y_positions = [0, 10]

ax.set_xticks(
    x_positions,
    ["0", "$10^{102}$", r"$10^{10^{10^{50}}}$", r"$\sim10^{10^{10^{50}}}$"],
    fontsize=12,
)
ax.set_yticks(y_positions, ["$10^{15}$", "$10^{60}$"], fontsize=12)

ax.scatter(x, y, color="tab:red", zorder=3, label=r"experimental data (n=1)")
ax.plot(x_line, y_line, color="black", linestyle="-.", linewidth=1, label=r"linear fit, $r^{2}=1.00000$")

marker_specs = [
    (10, "Black holes \n evaporate"),
    (80, "Baby universe \n is born"),
    (40, "Boltzmann brains' \n revolution"),
]

for x_pos, text in marker_specs:
    ax.axvline(x_pos, color="tab:blue", linestyle="--", linewidth=1, alpha=0.7)
    ax.annotate(
        text,
        xy=(x_pos, ax.get_ylim()[1]),
        xytext=(0, 5),
        textcoords="offset points",
        ha="center",
        va="bottom",
        fontsize=12,
        color="tab:blue",
    )


ax.set_xlabel(r"Elapsed time (years), hyperlog scale", fontsize=16)
ax.set_ylabel(r"$\mathrm{H_{2}O}$ yield (moles)", fontsize=16)
ax.legend()

plt.show()
```
